## 课程要点

使用代价函数的计算整体最优值适用于小数据量的场景，一次计算就可以算出最优值。但是对于，大数据、大模型的深度学习而言也会有很大的限制（硬件机器要非常的好），所以要有学习小批量数据更新权重的方式，也是最近比较常用的方式：梯度下降与反向传播。

```python
import dataset
from matplotlib import pyplot as plt
import numpy as np
```

```python
# 模拟数据集
data_size = 100
xs,ys = dataset.get_beans(data_size)
print(xs)
print(ys)
```

```python
# 绘制豆豆大小和毒性关系的图表
plt.title("Size - Toxicity", fontsize = 12)
plt.xlabel("Bean Size")
plt.ylabel("Toxicity")

# 绘制豆豆的数据
plt.scatter(xs,ys)

# 权重，先设置一个默认值
weight = 0.2
y_pre = weight * xs

# 绘制预测的数据
plt.plot(xs,y_pre)
plt.show()
```

```python
# 梯度下降法求解最优值

weights = []
for index in range(100):
    x = xs[index]
    y = ys[index]
    # a=X^2
    # b=-2*×*y 
    # #C=y^2
    # 斜率k=2aw+b
    k = 2 * x**2 * weight  + (-2*x*y)
    learn_rate = 0.2
    weight = weight - learn_rate * k
    weights.append(weight)

# 绘制 weight 的变化趋势
plt.title("Step - Weight", fontsize = 12)
plt.xlabel("Steps")
plt.ylabel("Weights")
plt.plot(weights)
plt.show()
```

```python
# 绘制学习到的 weight 是否可以正确预测豆豆的大小和毒性的关系

print(f"weight: {weight}")
plt.scatter(xs, ys)
y_pre = weight * xs
plt.plot(xs, y_pre)
plt.show()
```
