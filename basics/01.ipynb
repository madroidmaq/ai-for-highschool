## 课程要点：Rosenblatt 感知器

下面是一个Rosenblatt感知器的示意图：

1. 多个输入： x_1, x_2, ..., x_n
2. 每个输入都有一个对应的权重：w_1, w_2, ..., w_n
3. 所有的加权输入被累加，然后加上偏置： b 
4. 结果通过一个激活函数（如阶跃函数）进行处理，产生输出： y
